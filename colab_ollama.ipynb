```python
# Cell 6: Start the SEO app
import subprocess, time

proc = subprocess.Popen(
    ['python', '-m', 'uvicorn', 'main:app', '--host', '0.0.0.0', '--port', '8000'],
    cwd='/content/frog',
    stdout=subprocess.PIPE, stderr=subprocess.STDOUT
)
time.sleep(6)
out = proc.stdout.read1(4096).decode()
if 'running' in out.lower() or 'started' in out.lower():
    print('✅ SEO App is running on port 8000!')
else:
    print('Output:', out[:500])
```
